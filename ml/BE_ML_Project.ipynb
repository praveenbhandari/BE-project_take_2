```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import random as r
```

```python
data=pd.read_csv("train.csv", na_values=' ')
data
data.Dependents.unique()
data
# y
# data.isnull().sum().sort_values(ascending=False)
data = data.fillna(1)
```

```python
def enc(df):
    df.replace(to_replace="Male",value=1,inplace=True)
    df.replace(to_replace="Female",value=0,inplace=True)
    df.replace(to_replace="Yes",value=1,inplace=True)
    df.replace(to_replace="No",value=0,inplace=True)
    df.replace(to_replace="Urban",value=1,inplace=True)
    df.replace(to_replace="Rural",value=0,inplace=True)
    df.replace(to_replace="Semiurban",value=2,inplace=True)
    df.replace(to_replace="Graduate",value=1,inplace=True)
    df.replace(to_replace="Not Graduate",value=0,inplace=True)
    df.replace(to_replace="Y",value=1,inplace=True)
    df.replace(to_replace="N",value=0,inplace=True)
    df.replace(to_replace="3+",value=4,inplace=True)
#     df.replace(to_replace=" ",value=0,inplace=True)
    
enc(data)
# enc(y)
data
y=data["Loan_Status"]
# data.dropna()
data.drop("Loan_Status",axis=1,inplace=True)
data
```
